```python
import numpy as np 
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')


prevalence = 0.01 
sensitivity_test = 0.9
specificity_test = 0.91
fnr = 1. - sensitivity_test
fpr = 1. - 0.91 
```

```python
def get_posterior_prob(prior, sensitivity, specificity):
    
    return prior*sensitivity/(prior*sensitivity + (1-prior)*(1-specificity))
```

```python
from matplotlib.pyplot import figure
x = np.linspace(0.7, 1, 50)
y = np.linspace(0.7, 1, 50)

X, Y = np.meshgrid(x, y)
for p in np.linspace(0.1, 0.5, 11):
    Z = get_posterior_prob(p,X, Y)
    plt.figure()
    figure(figsize=(8, 6), dpi=80)
    plt.contourf(X, Y, Z, 20, cmap='seismic',vmin = 0, vmax=1)
    #contours = plt.contour(X, Y, Z, 3, colors='black')
    #plt.contour(X, Y, Z, 20, cmap='RdGy');
    #plt.colorbar();

    plt.xlabel("Test Sensitivity")
    plt.ylabel("Test Specificity")
    plt.title("""Probability of Being Positive Given Positive Test
                 at Prevalence Rate = {}
              """.format(round(p,2)))
    plt.savefig('{}.png'.format(round(p,2)))
    plt.colorbar()
```
